```python
import os
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text
```

```python
resume_texts = []
resume_names = []

for file in os.listdir("resumes"):
    if file.endswith(".txt"):
        with open("resumes/" + file, "r", encoding="utf-8") as f:
            text = f.read()
            resume_texts.append(clean_text(text))
            resume_names.append(file)

print("Resumes Loaded:")
print(resume_names)
```

```python
with open("job_description.txt", "r", encoding="utf-8") as f:
    job_desc = clean_text(f.read())

print("Job Description Loaded")
```

```python
with open("job_description.txt", "r", encoding="utf-8") as f:
    job_desc = clean_text(f.read())

print("Job Description Loaded Successfully")
```

```python
vectorizer = TfidfVectorizer()

all_documents = resume_texts + [job_desc]

tfidf_matrix = vectorizer.fit_transform(all_documents)

print("TF-IDF Matrix Created Successfully")
```

```python
similarity_scores = cosine_similarity(
    tfidf_matrix[-1],   # job description
    tfidf_matrix[:-1]   # resumes
)

scores = similarity_scores[0]

print("Similarity Scores:")
print(scores)
```

```python
ranking = pd.DataFrame({
    "Candidate": resume_names,
    "Similarity Score": scores
})

ranking = ranking.sort_values(by="Similarity Score", ascending=False)

ranking
```
